```python
import matplotlib.pyplot as plt
from matplotlib.patches import Arc
import numpy as np


# Euklidische Distanz
def plot_euclidean_distance(p1, p2, figsize=(8, 6), dpi=100):
    fig = plt.figure(figsize=figsize, dpi=dpi)
    ax = fig.add_subplot(111)
    ax.plot([p1[0], p2[0]], [p1[1], p2[1]], 'ro-', zorder=1)
    ax.scatter(p1[0], p1[1], c='blue', label='Punkt 1', zorder=2)
    ax.scatter(p2[0], p2[1], c='green', label='Punkt 2', zorder=2)
    # plt.legend()
    ax.set_title('Euklidische Distanz')
    ax.set_xlim(0, 5)
    ax.set_ylim(0, 4)
    ax.grid(False)
    ax.set_yticks(list(range(max(p1[0], p2[0])+1)))

    # Entferne die Rahmen rechts und oben
    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)

    ax.set_aspect('equal', adjustable='box')
    return fig



# Manhattan-Distanz
def plot_manhattan_distance(p1, p2, figsize=(8, 6), dpi=100):
    fig = plt.figure(figsize=figsize, dpi=dpi)
    ax = fig.add_subplot(111)
    ax.plot([p1[0], p2[0]], [p1[1], p1[1]], 'ro--', label='Horizontale Distanz', zorder=1)
    ax.plot([p2[0], p2[0]], [p1[1], p2[1]], 'ro--', label='Vertikale Distanz', zorder=1)
    ax.scatter(p1[0], p1[1], c='blue', label='Punkt 1', zorder=2)
    ax.scatter(p2[0], p2[1], c='green', label='Punkt 2', zorder=2)
    # plt.legend()
    ax.set_title('Manhattan-Distanz')
    ax.set_xlim(0, 5)
    ax.set_ylim(0, 4)
    ax.set_yticks(list(range(max(p1[0], p2[0])+1)))
    ax.set_aspect('equal', adjustable='box')

    # Entferne die Rahmen rechts und oben
    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)

    return fig


# Cosinus-Ähnlichkeit
def plot_cosine_similarity(p1, p2, figsize=(8, 6), dpi=100):
    origin = [0, 0]
    fig = plt.figure(figsize=figsize, dpi=dpi)
    ax = fig.add_subplot(111)
    ax.quiver(*origin, *p1, color='blue', angles='xy', scale_units='xy', scale=1, zorder=1, label='Vektor 1')
    ax.quiver(*origin, *p2, color='green', angles='xy', scale_units='xy', scale=1, zorder=1, label='Vektor 2')

    # Berechnen des Winkels für den Arc
    theta1 = np.arctan2(p1[1], p1[0])
    theta2 = np.arctan2(p2[1], p2[0])
    if theta1 > theta2:
        theta1, theta2 = theta2, theta1
        p1, p2 = p2, p1

    angle = np.degrees(theta2 - theta1)

    # Zeichnen des Arcs
    arc = Arc(origin, 2, 2, angle=np.degrees(theta1), theta1=0, theta2=angle, color='red', label='Winkel')
    ax.add_patch(arc)

    ax.set_aspect('equal', adjustable='box')
    ax.set_xlim(0, 5)
    ax.set_ylim(0, 4)
    # plt.legend()
    ax.set_yticks(list(range(max(p1[0], p2[0])+1)))
    ax.set_title('Cosinus-Ähnlichkeit')

    # Entferne die Rahmen rechts und oben
    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)

    return fig



# Punkte für die Visualisierung
point1 = np.array([1, 2])
point2 = np.array([4, 3])

figsize = (4, 3)
dpi = 100

# Plot all
eucl = plot_euclidean_distance(point1, point2, figsize=figsize, dpi=dpi)
eucl.savefig("images/euclidean_distance.jpg")

manh = plot_manhattan_distance(point1, point2, figsize=figsize, dpi=dpi)
manh.savefig("images/manhattan_distance.jpg")

cosn = plot_cosine_similarity(point1, point2, figsize=figsize, dpi=dpi)
cosn.savefig("images/cosine_similarity.jpg")
# plot_manhattan_distance(point1, point2)
# plt.savefig("manhatten_distance.jpg")
# plot_cosine_similarity(point1, point2)
# plt.savefig("cosine_similarity.jpg")
```

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.patches import Arc

def plot_cosine_similarity(p1, p2):
    origin = [0, 0]
    plt.figure()
    plt.quiver(*origin, *p1, color='blue', angles='xy', scale_units='xy', scale=1, zorder=1, label='Vektor 1')
    plt.quiver(*origin, *p2, color='green', angles='xy', scale_units='xy', scale=1, zorder=1, label='Vektor 2')

    # Berechnen des Winkels für den Arc
    theta1 = np.arctan2(p1[1], p1[0])
    theta2 = np.arctan2(p2[1], p2[0])
    if theta1 > theta2:
        theta1, theta2 = theta2, theta1
        p1, p2 = p2, p1

    angle = np.degrees(theta2 - theta1)

    # Zeichnen des Arcs
    arc = Arc(origin, 2, 2, angle=np.degrees(theta1), theta1=0, theta2=angle, color='red', label='Winkel')
    plt.gca().add_patch(arc)

    plt.gca().set_aspect('equal', adjustable='box')
    plt.xlim(0, 5)
    plt.ylim(0, 5)

    plt.legend()
    plt.title('Cosinus-Ähnlichkeit')
    plt.show()

# Beispielvektoren
point1 = np.array([1, 2])
point2 = np.array([4, 3])
plot_cosine_similarity(point1, point2)
```
